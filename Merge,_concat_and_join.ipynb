```python
data1 = {
    'Name':['Parshva', 'raj', 'rajat','sayad','rahul'],
    'age':['21', '20','25','30','21']
}

data2 = {
    'Name':['Parshva', 'raj', 'rajat','sayad','rahul'],
    'marks':['100', '80','75','90','40']
}
```

```python
import pandas as pd
```

```python
df1 = pd.DataFrame(data1)
```

```python
df2 = pd.DataFrame(data2)
```

```python
# df2.iloc[0,0] = 'parshva'
```

```python
dfconcat = pd.concat([df1, df2], axis=1, keys=['x','y'])
```

```python
dfconcat
```

```python
# df1 = df1.rename(columns = {'Name':'new_name','age':'new_age'})
```

```python
dfmerge = pd.merge(df1, df2, on='Name', how='left')
```

```python
dfmerge
```

```python
dfmerge1 = df1.merge(df2, on='Name', how='inner',indicator =True)
```

```python

dfmerge1
```

```python
dfjoin = df2.join(df1,how='inner',lsuffix='df2', rsuffix='df1')
```
