```python
data1 = {
    'Name':['Parshva', 'raj', 'rajat','sayad','rahul'],
    'age':['21', '20','25','30','21']
}

data2 = {
    'Name':['Parshva', 'raj', 'rajat','sayad','rahul'],
    'marks':['100', '80','75','90','40']
}
```

```python
import pandas as pd
```

```python
df1 = pd.DataFrame(data1)
```

```python
df2 = pd.DataFrame(data2)
```

```python
# df2.iloc[0,0] = 'parshva'
```

```python
dfconcat = pd.concat([df1, df2], axis=1, keys=['x','y'])
```

```python
dfconcat
```

```python
dfmerge = pd.merge(df1, df2,on='Name', how='right')
```
